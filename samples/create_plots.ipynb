```python
# import libraries

from openpile.construct import Pile, SoilProfile, Model, Layer
import matplotlib.pyplot as plt
from pathlib import Path

# link to folder where plots should be stored
folder = Path.cwd().parent / "docs" / "source" / "_static" / "plots"
```

```python
# save the figure generated by Pile.plot()
p = Pile(
    name="<pile name>",
    material="Steel",
    kind="Circular",
    top_elevation=10,
    pile_sections={
        "length": [10, 10, 10],
        "wall thickness": [0.04, 0.06, 0.08],
        "diameter": [4.0, 4.0, 4.0],
    },
)

p.plot()
plt.savefig(
    folder / "Pile_plot.png",
)
```

```python
# save the figure generated by SoilProfile.plot()
sp = SoilProfile(
    name="<soil profile name>",
    top_elevation=0,
    water_line=8.0,
    layers=[
        Layer(name="<layer name>", top=0, bottom=-4, weight=18),
        Layer(name="<layer name>", top=-4, bottom=-8, weight=18),
        Layer(name="<layer name>", top=-8, bottom=-20, weight=18),
        Layer(name="<layer name>", top=-20, bottom=-30, weight=18),
    ],
)
sp.plot()
plt.savefig(
    folder / "SoilProfile_plot.png",
)
```

```python
# save the figure generated by Model.plot()

m = Model(name="<Model name>", pile=p)
m.set_pointload(elevation=0, Py=-500)
m.set_support(elevation=10, Tx=True, Ty=True)
m.set_support(elevation=-20, Ty=True)
m.plot()
plt.savefig(
    folder / "Model_no_soil_plot.png",
)
```

```python
# save the figure generated by Model.plot()

m = Model(name="<Model name>", pile=p, soil=sp)
m.set_pointload(elevation=10, Py=500)
m.plot()
plt.savefig(
    folder / "Model_with_soil_plot.png",
)
```
